```python
for i in num_cols:
    q1 = df[i].quantile(0.25)
    q2 = df[i].quantile(0.75)
    iqr = q2-q1
    lb = q1-iqr*1.5
    ub = q2+iqr*1.5
    df = df[(df[i]>=lb) & (df[i]<=ub )]

```
